```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_data = cv2.imread("coins.jpg")
```

```python
image = cv2.cvtColor(image_data, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image_data, cv2.COLOR_BGR2GRAY)
blur_ = cv2.GaussianBlur(gray, (11,11), 0)

edge_ = cv2.Canny(blur_, 30, 250)
plt.imshow(edge_)
plt.show()
```

```python
cnt , hierarchy = cv2.findContours(edge_.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnt_image = image_data.copy()
cv2.drawContours(cnt_image, cnt, -1, (0,255,0), 2)
plt.imshow(cnt_image)
plt.axis('off')
plt.show()
```
